```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("Rajasthan_Heatwave.csv")
data.head()
```

```python
data.info()
```

```python
data.columns
```

```python
data['HEATWAVE'].value_counts()
```

```python
data.isnull().sum()
```

```python
sns.scatterplot(x=data['WIND_U10'],y=data['WIND_V10'])
plt.xlabel("10m U-component of Wind ")
plt.ylabel("10m V-component of wind")
plt.title("Wind U10 VS Wind V10")
plt.show()
```

```python
sns.histplot(x=data['WIND_U10'],kde=True)
plt.xlabel("10m U-component of wind")
plt.ylabel("count")
plt.title("Wind U-component")
plt.show()
```

```python
sns.histplot(x=data["WIND_V10"],kde=True)
plt.xlabel("10m V-component of wind")
plt.ylabel("Count")
plt.title("Wind V-component")
plt.show()
```

```python
sns.histplot(x=data["MSLP"],kde=True)
plt.xlabel("Mean Sea Level Pressure")
plt.ylabel("Count")
plt.title("Mean sean level pressure(MSLP)")
plt.show()
```

```python
sns.histplot(x=data["BLH"],kde=True)
plt.xlabel("Boundary Level Height")
plt.ylabel("Count")
plt.title("Boundary Level Height")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x=data["GEOP"],hue=data["HEATWAVE"])
plt.xlabel("Geopotential")
plt.ylabel("count")
plt.title("Geopotential")
plt.xticks(rotation=90)
plt.show()
```

```python
sns.histplot(x=data["TEMP2M"],kde=True)
plt.xlabel("2m temperature")
plt.ylabel("count")
plt.title("2m Temperature")
plt.show()
```

```python
sns.histplot(x=data["TMIN"],kde=True)
plt.xlabel("Minimum 2m Temperature")
plt.ylabel("Count")
plt.title("Minimum 2m Temperature")
plt.show()
```

```python
sns.histplot(x=data["DEW2M"],kde=True)
plt.title("2m dew point temperature")
plt.xlabel("2m dew point temperature")
plt.ylabel("count")
plt.show()
```
